**4. In a markdown block, outline why Linear Regression is not a good machine learning model for classification.**

Linear regression isn't the best model for classification as usually for classification, you are assigning either one class or the other based on the input value. Assigning a straight line to classify data creates a problem in the midpoint where the class changes, where a small change in the value can completely change the class, creating indecision.

__5. In a markdown block, outline the machine learning Logistic Regression algorithm.__

Logistic regression is a binary classification algorithm that predicts the probability of a value being a class. It works by drawing a sigmoid function (s shaped curve that goes from 0 to 1) and mapping that curve to the probability of an input value being a certain class. The actual training of a logistic regression model doesn't modify the sigmoid curve, but it changes the decision boundary, which is the border between being one class or the other.
